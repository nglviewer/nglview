```python
import nglview as nv
from nglview.contrib.movie import MovieMaker
import simpletraj

class FiveFrameTrajectory(nv.SimpletrajTrajectory):
    @property
    def n_frames(self):
        return 5
        
traj = FiveFrameTrajectory(nv.datafiles.XTC, nv.datafiles.PDB)
view = nv.show_simpletraj(traj)
view.center()
view.add_surface('protein', opacity=0.5, color='residueindex')
view.control.orient([15.69986420582941,
 114.8276262266868,
 48.816742057652796,
 0,
 -18.22863682402078,
 -46.5629522451937,
 115.38869989652724,
 0,
 123.43490126182644,
 -21.481422658119275,
 10.831316500165215,
 0,
 -39.1268284890941,
 -46.69809684423501,
 -44.63013982367478,
 1])
view
```

```python
# require: `pip install moviepy`

movie = MovieMaker(view, in_memory=True, output='abc.gif')
movie.make()
```
