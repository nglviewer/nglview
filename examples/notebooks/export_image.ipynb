````python
import nglview as nv

view = nv.show_structure_file(nv.datafiles.PDB)
view
````

# Save and restore orientation

To reproduce exact figures it is helpful to save the orientation. This can be done by activating
the camera sync and get/set the orientation

````python
view._set_sync_camera([view])
````

````python
# get orientation after rotation of molecule
view._camera_orientation
````

````python
# set orientation 
# this will update the orientation but must be executed in a different cell
view._set_camera_orientation([-102.85544137184395,
 -4.572939454457459,
 -80.02581518067477,
 0,
 -2.2884601860244054,
 130.3025560982287,
 -4.504610374651535,
 0,
 80.123690757245,
 -2.1486715422100775,
 -102.85845653335238,
 0,
 -37.284461975097656,
 -42.47041130065918,
 -44.55349826812744,
 1])
````
